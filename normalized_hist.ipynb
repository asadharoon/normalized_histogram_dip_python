<h1 align="center">Analysis of Image using Histogram in Python</h1>

# Import Libraries

```python
from skimage import data
from scipy.misc import imread, imresize
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

# Original Image

```python
original_image =imread('low_contrast_image_histogram.png',False, 'L')    #read image as grey scale image
```

# Processed Image

```python
processed_img =original_image.copy()
plt.imshow(processed_img,cmap=plt.cm.gray)
```

# Finding Probabilities

```python
greylevels, counts = np.unique(processed_img, return_counts=True)
prob=counts/sum(counts)
x,y=greylevels,prob
```

# Normalized Histogram

```python
plt.figure(figsize=(8,8))
plt.bar(x,y,align='edge',width=0.7)
plt.show()
```

# Simple Histogram

```python
plt.figure(figsize=(8,8))
plt.hist(processed_img.flatten(),bins=256,range=[0,255],width=0.5)
plt.show()
```
